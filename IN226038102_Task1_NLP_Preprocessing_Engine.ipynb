Task 2:Build Advanced Preprocessing Function


```python
import re
import numpy as np
from collections import Counter
```

```python
import re

def preprocess_text(text):
    # Handle invalid input
    if not text or not isinstance(text, str):
        return []

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove punctuation
    text = re.sub(r'[^\w\s]', ' ', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Tokenization
    tokens = text.split()

    # 8. Remove very short words (≤2) except "no", "not"
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ['no', 'not']
    ]

    return cleaned_tokens


# Example test
text = "WOW!!! This is soooo goooddd!!! Visit http://example.com now. I have 2 dogs!! not bad"
output = preprocess_text(text)

print(output)
```

Task 3: Stress Testing


```python
# Function to convert tokens back to sentence
def tokens_to_sentence(tokens):
    return " ".join(tokens)

# Stress Test Cases
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Run tests
for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)
    cleaned_sentence = tokens_to_sentence(tokens)

    print(f"Test Case {i}:")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 60)
```

Task 4: Token Analytics


```python
# Function to calculate token analytics
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, round(avg_token_length, 2)


# Reuse previous test sentences
for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)
    cleaned_sentence = " ".join(tokens)

    total, unique, avg_len = token_analytics(tokens)

    print(f"Test Case {i}:")
    print("Original Text      :", sentence)
    print("Cleaned Tokens     :", tokens)
    print("Cleaned Sentence   :", cleaned_sentence)
    print("Total Tokens       :", total)
    print("Unique Tokens      :", unique)
    print("Avg Token Length   :", avg_len)
    print("-" * 60)
```

A: Most meaningful tokens: "I am not happy with this"  retained the most meaningful tokens because it contained clean, structured text with minimal noise, and important words like "not" were preserved.


Task 5: Frequency Analysis

```python
from collections import Counter

# Combine all tokens
all_tokens = []
for sentence in test_sentences:
    all_tokens += preprocess_text(sentence)

# Count frequency
freq = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
for word, count in freq.most_common(10):
    print(word, ":", count)

# Top 5 least frequent words (simple way)
print("\nTop 5 Least Frequent Words:")
least_5 = sorted(freq.items(), key=lambda x: x[1])[:5]
for word, count in least_5:
    print(word, ":", count)
```

Task 6: Build Full Pipeline


```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Example usage
result = full_pipeline(test_sentences)

print("Tokens:", result["tokens"])
print("\nClean Sentences:", result["clean_sentences"])
```
